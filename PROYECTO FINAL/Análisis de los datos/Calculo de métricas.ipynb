## NORMALIZACIÓN DE DATOS:

```python
columnas_tiros = ['Goles', '% Disparos a puerta', 'Total de tiros/90', 'Disparos a puerta/90', 'Goles/tiros', 'Goles/tiros a puerta']
df_selected = estad_jugador_final[columnas_tiros]

scaler = MinMaxScaler()
df_normalized = pd.DataFrame(scaler.fit_transform(df_selected), columns=columnas_tiros)

estad_jugador_final['Métrica de Tiro'] = df_normalized.sum(axis=1)

metrica_tiros = estad_jugador_final[['Jugador', 'Métrica de Tiro']]
metrica_tiros
```

```python
columnas_defensa = ['Derribos', 'Derribos completados', 'Regateadores tackleados', 'Bloqueos', 'Disparos bloqueados', 
                  'Pases bloqueados', 'Jug. tackleados + intercepciones', 'Despejes', 'Pelotas recuperadas', 
                  '% Duelos aéreos ganados']

df_selected = estad_jugador_final[columnas_defensa]

scaler = MinMaxScaler()
df_normalized = pd.DataFrame(scaler.fit_transform(df_selected), columns=columnas_defensa)

estad_jugador_final['Métrica de Defensa'] = df_normalized.sum(axis=1)

metrica_defensa = estad_jugador_final[['Jugador', 'Métrica de Defensa']]
metrica_defensa
```

```python
columnas_pase = ['Pases profundos', 'Pases completados', 'Pases intentados', '% Pases completados', 
                    'Pases cortos completados', 'Pases cortos intentados', '% Pases cortos completados', 
                    'Pases medios completados', 'Pases medios intentados', '% Pases medios completados', 
                    'Pases largos completados', 'Pases largos intentados', '% Pases largos completados']

df_selected = estad_jugador_final[columnas_pase]

scaler = MinMaxScaler()
df_normalized = pd.DataFrame(scaler.fit_transform(df_selected), columns=columnas_pase)

estad_jugador_final['Métrica de Pase'] = df_normalized.sum(axis=1)

metrica_pase = estad_jugador_final[['Jugador', 'Métrica de Pase']]
metrica_pase
```

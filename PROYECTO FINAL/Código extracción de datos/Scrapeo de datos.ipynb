```python
players_data_laliga = players_data_laliga.drop(columns = ['Born', '90s', 'SCA',
       'SCA90', 'PassLive', 'PassDead', 'TO', 'Sh', 'Fld', 'Def', 'GCA',
       'GCA90', 'PassLive', 'PassDead', 'TO', 'Sh', 'Fld', 'Def'])
players_data_laliga
```
